```python
# Define the function
def f(x):
    return 2*x**3 + 4*x**2 - 6*x + 1

# Trapezoidal Rule
a = 1
b = 5
h = 1

# Calculate f(x) at each point
x0 = f(1)
x1 = f(2)
x2 = f(3)
x3 = f(4)
x4 = f(5)

# Apply trapezoidal rule
trap = (h / 2) * (x0 + 2*x1 + 2*x2 + 2*x3 + x4)

# Calculate exact integral manually
# From antiderivative: x^4/2 + (4x^3)/3 - 3x^2 + x
def exact(x):
    return (x**4)/2 + (4*x**3)/3 - 3*x**2 + x

exact_value = exact(5) - exact(1)

# Calculate percentage error
error = abs(trap - exact_value) / exact_value * 100

# Print results
print("Trapezoidal Approximation =", trap)
print("Exact Value =", round(exact_value, 2))
print("Percentage Error =", round(error, 2), "%")
```
