```python
import numpy as np
from scipy.interpolate import lagrange

# Given data
t = np.array([0, 2, 5, 9, 15])
h = np.array([0.0, 12.1, 65.5, 197.6, 499.0])

# Lagrange interpolation
poly = lagrange(t, h)

# Get polynomial coefficients (highest degree first)
coeffs = poly.coefficients

# First derivative (velocity)
velocity_coeffs = np.polyder(coeffs)

# Second derivative (acceleration)
acceleration_coeffs = np.polyder(velocity_coeffs)

# Evaluate at t = 7
t_val = 7
velocity = np.polyval(velocity_coeffs, t_val)
acceleration = np.polyval(acceleration_coeffs, t_val)

# Print results
print("Velocity at t = 7 seconds:", round(velocity, 2), "m/s")
print("Acceleration at t = 7 seconds:", round(acceleration, 2), "m/s²")
```
