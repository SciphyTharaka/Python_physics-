```python
# Define the function
def f(x):
    return 4*x**4 + 3*x**3 - 2*x**2 + 5*x - 7

# Simpson's 1/3 Rule for 3 points: x0=0, x1=1, x2=2
x0 = 0
x1 = 1
x2 = 2
h = 1  # (x2 - x0) / 2

# Apply Simpson's Rule
simpson = (h / 3) * (f(x0) + 4*f(x1) + f(x2))


def exact_integral(x):
    return (4*x**5)/5 + (3*x**4)/4 - (2*x**3)/3 + (5*x**2)/2 - 7*x

exact_value = exact_integral(2) - exact_integral(0)


error = abs(simpson - exact_value) / exact_value * 100


print("Simpson's Approximation =", round(simpson, 2))
print("Exact Value =", round(exact_value, 2))
print("Percentage Error =", round(error, 2), "%")
```
