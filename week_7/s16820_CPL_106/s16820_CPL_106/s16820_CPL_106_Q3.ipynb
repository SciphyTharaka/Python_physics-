```python
import numpy as np
import matplotlib.pyplot as plt

Fs = 1000  # Sampling rate (samples per second)
T = 1      # Duration in seconds
N = Fs * T  # Total number of samples

t = np.linspace(0, T, N, endpoint=False)  # Time vector


signal = 2 * np.sin(2 * np.pi * 100 * t) + 1 * np.sin(2 * np.pi * 200 * t) # Create the composite signal

# Apply FFT
fft_result = np.fft.fft(signal)
frequencies = np.fft.fftfreq(N, 1/Fs)  # Frequency bins

# Get the magnitude spectrum
magnitude = np.abs(fft_result) / N  # Normalize

# Only keep the positive frequencies
half_N = N // 2
frequencies = frequencies[:half_N]
magnitude = magnitude[:half_N]


plt.figure(figsize=(10, 4))
plt.plot(frequencies, magnitude)
plt.title('FFT of Composite Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 0.1, 1000)  # 1000 samples in [0, 0.1]

Ac = 5         # Amplitude of carrier
Am = 2         # Amplitude of message signal
fc = 500       # Carrier frequency in Hz
fm = 50        # Modulating frequency in Hz

# Carrier wave
yc = Ac * np.sin(2 * np.pi * fc * t)

# Modulating signal (message signal)
ym = Am * np.sin(2 * np.pi * fm * t)

#  Amplitude Modulated (AM) signal
y = (Ac + ym) * np.sin(2 * np.pi * fc * t)


plt.figure(figsize=(12, 6))
plt.plot(t, yc, label='Carrier signal $y_c(t)$', alpha=0.6)
plt.plot(t, ym, label='Message signal $y_m(t)$', alpha=0.8)
plt.plot(t, y, label='AM signal $y(t)$', linewidth=2)

plt.xlabel('Time [s]')
plt.ylabel('Amplitude [V]')
plt.title('Carrier, Message, and AM Signals')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


Fs = 10000  # Sampling frequency (10 kHz)
T = 1 / Fs
t = np.linspace(0, 0.1, int(Fs * 0.1), endpoint=False)  # t in [0, 0.1)

# Signal parameters
Ac = 5         # Carrier amplitude
Am = 2         # Message amplitude
fc = 500       # Carrier frequency
fm = 50        # Message frequency

# Signals
ym = Am * np.sin(2 * np.pi * fm * t)
y = (Ac + ym) * np.sin(2 * np.pi * fc * t)

# FFT
N = len(t)
Y_fft = np.fft.fft(y)
freq = np.fft.fftfreq(N, T)


half_N = N // 2
Y_magnitude = np.abs(Y_fft[:half_N]) * 2 / N
frequencies = freq[:half_N]


plt.figure(figsize=(10, 5))
plt.plot(frequencies, Y_magnitude)
plt.title("Fourier Spectrum of AM Signal")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.xlim(0, 1000)  # Zoom to relevant frequencies
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

Fs = 10000  # Sampling frequency (Hz)
T = 1 / Fs
t = np.linspace(0, 0.1, int(Fs * 0.1), endpoint=False)  # t from 0 to 0.1 s

ym = 10 * np.sin(200 * np.pi * t) + 40 * np.sin(100 * np.pi * t)  # y_m(t)
yc = 20 * np.sin(400 * np.pi * t)  # y_c(t)
y = (20 + ym) * np.sin(2 * np.pi * 200 * t)  # AM signal y(t)

plt.figure(figsize=(12, 6))
plt.plot(t, ym, label='Modulating $y_m(t)$', color='blue')
plt.plot(t, yc, label='Carrier $y_c(t)$', color='orange')
plt.plot(t, y, label='Modulated $y(t)$', color='green')
plt.title("Amplitude Modulation with Multi-Frequency Signal - Time Domain")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sampling configuration
Fs = 10000  # Sampling frequency (Hz)
T = 1 / Fs
t = np.linspace(0, 0.1, int(Fs * 0.1), endpoint=False)

# Modulating and carrier signals (same as Part b)
ym = 10 * np.sin(200 * np.pi * t) + 40 * np.sin(100 * np.pi * t)
y = (20 + ym) * np.sin(2 * np.pi * 200 * t)  # AM signal y(t)

# FFT
N = len(t)
Y_fft = np.fft.fft(y)
freqs = np.fft.fftfreq(N, T)

# Positive frequencies only
half_N = N // 2
Y_magnitude = np.abs(Y_fft[:half_N]) * 2 / N
positive_freqs = freqs[:half_N]

# Plot FFT spectrum
plt.figure(figsize=(10, 5))
plt.plot(positive_freqs, Y_magnitude)
plt.title("FFT of Modulated Signal with Complex Modulation (Part IV)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.xlim(0, 1000)
plt.tight_layout()
plt.show()

```
