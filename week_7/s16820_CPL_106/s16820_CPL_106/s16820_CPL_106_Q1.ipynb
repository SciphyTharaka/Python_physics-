```python
import numpy as np
import matplotlib.pyplot as plt

# Given parameters
A = 10  # amplitude in cm
F = 50  # frequency in Hz
pha = np.pi / 2  # phase in radians

# Signal function
def signal(t):
    return A * np.sin(2 * np.pi * F * t + pha)
t_end = 0.05
t_cont = np.linspace(0, t_end, 10000)  # x-axis

# Sampling frequencies
sampling_freqs = [10, 100, 1000, 10000]

# Plot sampled signals
for fs in sampling_freqs:
    t_samples = np.arange(0, t_end, 1/fs)  # X values for scatter 
    y_samples = signal(t_samples)    # y values for scatter 
    plt.scatter(t_samples, y_samples, label=f'Sampled at {fs} Hz', s=30)

plt.title('Sampling of a 50 Hz sinewave with different sampling rates')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude (cm)')
plt.legend()
plt.grid(True)
plt.show()
```
