# Q3 Polynomial Regression

# a) best degree 


```python
import numpy as np
import matplotlib.pyplot as plt
p= np.array([1.3,2.7,4.9,8.1,12.5,18.9,27.3,37.7,50.1,64.5])
u=np.linspace(0.2,2.0,10)
arr=[]
for i in range (0,9):
    a=i+1
    coe =np.polyfit(u,p,a)
    fit = np.polyval(coe,u)
    mse=np.mean((p-fit)**2)
    print("MSE ",a,"=",mse)
    arr.append(mse)
print("Best degree of polynomial for the above data : ",(np.argmin(arr)+1))
```

# b) polynomial curves 


```python
import numpy as np
import matplotlib.pyplot as plt
p= np.array([1.3,2.7,4.9,8.1,12.5,18.9,27.3,37.7,50.1,64.5])
u=np.linspace(0.2,2.0,10)


coe =np.polyfit(u,p,1)
fit_1 = np.polyval(coe,u)
plt.plot(u,p, 'o',color='r', label ='Original data line', markersize=5)
plt.plot(u,fit_1,color ='g',label='Fitted line when degree=1')

coe =np.polyfit(u,p,2)
fit_2 = np.polyval(coe,u)
plt.plot(u,fit_2,color ='yellow',label='Fitted line when degree=2')

coe =np.polyfit(u,p,3)
fit_3 = np.polyval(coe,u)
plt.plot(u,fit_3,color ='pink',label='Fitted line when degree=3')

coe =np.polyfit(u,p,7)
fit_7 = np.polyval(coe,u)
plt.plot(u,fit_7,color ='orange',label='Fitted line when degree=7')

coe =np.polyfit(u,p,10)
fit_10 = np.polyval(coe,u)
plt.plot(u,fit_10,color ='darkblue',label='Fitted line when degree=10')

plt.xlabel("velocity of the particle(u)")
plt.ylabel(" momentum (p) ")
plt.legend()
plt.grid(True)
plt.show()
```

# c) Equation of best fit 


```python
p= np.array([1.3,2.7,4.9,8.1,12.5,18.9,27.3,37.7,50.1,64.5])
u=np.linspace(0.2,2.0,10)

# best fit degree is 10 
coe =np.polyfit(u,p,10)
pol=np.poly1d(coe) # equation of best fitted line 
print(coe)
print (pol)
```
