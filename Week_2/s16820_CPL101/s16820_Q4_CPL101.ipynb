```python
 import numpy as np 
 from scipy.interpolate import lagrange 
import matplotlib.pyplot as plt
 t = np.array([0, 1, 2,3,4,5])
 h = np.array([0, 5, 20, 45, 80, 125]) 
polynomial = lagrange(t, h)
 y1 = polynomial(t)

plt.scatter(t,h,label='Data Points',color='g')
plt.plot(t,y1,color='blue',label='predictor line')
plt.xlabel("time")
plt.ylabel("height")
plt.legend()
plt.grid(True)
plt.show()
```
