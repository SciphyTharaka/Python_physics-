# Regression and Interpolation

# Q2 Least Squares Regression

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1,1.9,10) # 0.1 to 1.9 10 steps 
y = np.array([1.4,2.4,3.1,3.7,4.2,4.6,5.1,5.4,5.7,6.14]) # data points

T2= y**2
print("Time square values are:",T2)

coe =np.polyfit(x,T2,1) # np.polyfit(x, y, deg) , it gives [m, c], computes slope & intercept of best-fit line
fit = np.polyval(coe,x) # uses that line to calculate fitted values, new y values 

plt.plot(x,fit,color ='g',label='Fitted line')
plt.plot(x,T2, 'o',color='r', label ='Original data line', markersize=8)
plt.legend()
plt.grid(True)
plt.show()

k = (((2*(np.pi))**2)/T2)*x
print("Values for sprin constant are:",k)
k1 =np.mean(k)
print("sprin constant (k) : ",k1)
```
