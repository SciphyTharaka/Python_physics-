```python
def bisection(f, a, b, tol):
    if f(a) * f(b) >= 0:
       print("The function must have opposite signs at a and b.")

    while (b - a) / 2 > tol:
        c = (a + b) / 2
        if f(c) == 0:
            return c

        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
            
    return (a + b) / 2
```

```python
import numpy as np

def f(x):
    return x**2 - np.exp(-x) - 2 * np.sin(x)

def bisection(f, a, b, tol): # f-fuction, a/b - interval, tol - error 
    if f(a) * f(b) >= 0:
        print("The function must have opposite signs at a and b.")
        return None

    while (b - a) / 2 > tol:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2


root = bisection(f, 0, 2, 1e-6)
print("Root of f(x): ", root)
```

```python
import numpy as np

def f(t):
    return 10 * t - 4.9 * t**2

# Bisection method
def bisection(f, a, b, tol): 
    if f(a) * f(b) >= 0:
        print("Bisection method fails. f(a) and f(b) must have opposite signs.")
        return None

    while (b - a) / 2 > tol:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

root = bisection(f, 0.1, 3, 1e-6)
print("Time when projectile returns to initial height:", root, "seconds")
```
