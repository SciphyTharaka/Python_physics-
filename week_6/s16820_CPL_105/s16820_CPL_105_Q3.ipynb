```python
# Q3 I. a) # Use the fsolve() function in Scipy Python package
```

```python
import numpy as np
from scipy.optimize import fsolve

def f(x):
    return x**2 - np.exp(-x) - 2 * np.sin(x)

initial_guess = 1.0
root = fsolve(f, initial_guess)

print("Root of f(x):", root[0]) #The fsolve() function always returns a NumPy array.

```

```python
import numpy as np
from scipy.optimize import fsolve

def f(t):
    return 10 * t - 4.9 * t**2

initial_guess = 2.0
root = fsolve(f, initial_guess)

print("Time when projectile returns to initial height:", root[0], "seconds")
```

```python
import numpy as np
from scipy.optimize import fsolve

def f(L):
    return 2 * np.pi * np.sqrt(L / 9.8) - 10

initial_guess = 50
root = fsolve(f, initial_guess)

print("Length L for T = 10 seconds:", root[0], "meters")
```

```python
import numpy as np
from scipy.optimize import fsolve

def f(t):
    return 5 * (1 - np.exp(-t / 10)) - 2.5

initial_guess = 10
time = fsolve(f, initial_guess)

print("Time at which Vc(t) = 2.5 V:", time[0], "seconds")
```
