```python
import numpy as np
k = 8.988e9  # 1 / (4 * pi * epsilon)
charges = np.array([2e-6, -4e-6, 6e-6])
positions = np.array([[1, 0],[0, 2], [-2, -1]])
Electric_filed = np.array([0.0, 0.0])
print("Electric fileds :")
for i in range (0,3):
    q = charges[i]
    r = np.linalg.norm(positions[i])  # distance form charge to origin or magnitude of vector
    vector = -positions[i]
    unit_vector= vector/r
    E= ((k*q)/(r**2))*unit_vector
    print (E) 
    Electric_filed += E
print("\n")
print("Total Electric filed of origin : \n",Electric_filed)
print("\n")
magnitude = np.linalg.norm(Electric_filed)
angle = np.degrees(np.arctan2(Electric_filed[1], Electric_filed[0]))  # angle from x-axis

print(" Magnitude of Total Electric Field :\n ",magnitude)
print("\n")
print(" Direction (angle from x-axis in degrees):\n",angle)
print("\n")
unit_vector = Electric_filed / magnitude
print("Unit Vector of Total Electric Field:\n", unit_vector)
    
```

```python
import numpy as np

q = 2e-6 
v = np.array([3, 4, 0])      
B = np.array([0, 0.2, 0.3])  


F = q * np.cross(v, B)

print("Magnetic Force Vector (N):", F)
```
