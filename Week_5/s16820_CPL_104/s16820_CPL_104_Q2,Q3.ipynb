```python
import numpy as np

def find_eigenvalues_and_eigenvectors(matrix):
    A = np.array(matrix)
    if A.shape[0] != A.shape[1]:
        raise ValueError("Matrix must be square")
    eigenvalues, eigenvectors = np.linalg.eig(A)
    return eigenvalues, eigenvectors

# Given values
mass = 10  # in kg
radius = 0.02  # in meters

# Moment of inertia for a solid sphere: (2/5) * M * R^2
I_value = (2/5) * mass * (radius ** 2)

# Inertia tensor (diagonal since sphere is symmetric)
I_tensor = [
    [I_value, 0, 0],
    [0, I_value, 0],
    [0, 0, I_value]
]

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = find_eigenvalues_and_eigenvectors(I_tensor)

# Display results
print("Moment of Inertia Tensor:")
print(np.array(I_tensor))

print("\nEigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)
```

```python
import numpy as np

def find_eigenvalues_and_eigenvectors(matrix):
    A = np.array(matrix)
    if A.shape[0] != A.shape[1]:
        raise ValueError("Matrix must be square")
    eigenvalues, eigenvectors = np.linalg.eig(A)
    return eigenvalues, eigenvectors

n = 3
hbar_omega = 1
factor = hbar_omega / 2 * np.sqrt(2)

H = [[factor * (i + j) for j in range(1, n + 1)] for i in range(1, n + 1)]

eigenvalues, eigenvectors = find_eigenvalues_and_eigenvectors(H)

print("Hamiltonian matrix:")
print(np.array(H))

print("\nEigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)
```
