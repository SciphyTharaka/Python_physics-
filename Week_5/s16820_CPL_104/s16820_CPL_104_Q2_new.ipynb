```python
import numpy as np

def matrix_sum(A, B):
    return np.add(A, B)

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

result = matrix_sum(A, B)

print("Matrix A:")
print(A)


print("\nMatrix B:")
print(B)

print("\nA + B =")
print(result)
```

```python
import numpy as np

def is_symmetric(A):
    return np.array_equal(A, A.T)

A = np.array([[1, 2, 3],
              [2, 5, 6],
              [3, 6, 9]])

print("Matrix A:")
print(A)

if is_symmetric(A):
    print("\nMatrix A is symmetric.")
else:
    print("\nMatrix A is not symmetric.")
```

```python
import numpy as np

def multiply_matrices(A, B):
    if A.shape[1] != B.shape[0]:
        print("Cannot multiply: Columns of A must equal rows of B.")
        return None
    else:
        return np.dot(A, B)

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

result = multiply_matrices(A, B)

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

if result is not None:
    print("\nA × B =")
    print(result)
```
