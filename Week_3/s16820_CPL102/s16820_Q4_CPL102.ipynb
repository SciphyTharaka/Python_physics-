```python
import numpy as np

start = 0
end = 2
points = 2000

def simpson_rule(start, end, points):
    x = np.linspace(start, end, points)
    y = 2 * np.pi * x**2.5
    h = (end - start) / (points - 1)
    integral = (h/3) * (y[0] + 2 * sum(y[2:-1:2]) + 4 * sum(y[1::2]) + y[-1])
    return integral

calculated = simpson_rule(start, end, points)
actual = (2 * np.pi * end**3.5) / 3.5

print("Actual value:", actual)
print("Calculated value:", calculated)
print("Error:", actual - calculated)
```
