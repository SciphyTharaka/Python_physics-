```python
h = 2.3
x = 5.43


def f(x):
    return x**2 - 1

def central_diff(f, x: float, h: float, order: int):
    if order == 1:
        return (f(x + h) - f(x - h)) / (2 * h)
    elif order == 2:
        return (f(x + h) - 2 * f(x) + f(x - h)) / (h ** 2)
    else:
        return None

print("Central Difference 1st derivative:", central_diff(f, x, h, 1))
print("Central Difference 2nd derivative:", central_diff(f, x, h, 2))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=-1
A =0.02
h =0.001
w=100
s= (np.pi)/6
f= A*np.cos((w*x)+s)
order1 = 1
order2 = 2

def central_diff(f,x,h,order):
    
    fhp =A*np.cos((w*(x+h))+s)
    fhn =A*np.cos((w*(x-h))+s)
    
    if order == 2:
        f1 = (fhn - (2*f) +fhp ) /(h**2)
    elif order == 1:
        f1 = (fhp- fhn) /(2*h)
    else:
        print('enter valid order(1 or 2)')
    return f1


f1 = central_diff(f,x,h,order1)
#print velosity of t = x point)
print ('Speed of the particle in t =',x,' time is =',f1) 

f1 = central_diff(f,x,h,order2)
#print velosity of t = x point)
print ('Acceleration of the particle in t =',x,' time is =',f1)

t=np.linspace(-4,4,8000)
X= A*np.cos((w*t)+s)
plt.plot(t, X, color="green", label="Simple Harmonic motion")
plt.legend()
plt.show()

X1= central_diff(f,t,h,order1)
plt.plot(t, X1, color="blue", label="Speed of the particle")
plt.legend()
plt.show()

X1= central_diff(f,t,h,order2)
plt.plot(t, X1, color="yellow", label="Acceleration of the particle")
plt.legend()
plt.show()



```
