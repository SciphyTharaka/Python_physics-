```python
h=2.3
x=5.43

def f(x):
    return x**2-1

def num_diff(f, x: float, h: float, method: str):
    if method == 'forward':
        return (f(x + h) - f(x)) / h
    else: 
        if method == 'backward':
            return (f(x) - f(x - h)) / h
        else: 
            if method == 'central':
                return (f(x + h) - f(x - h)) / (2 * h)
            else :
                print ("Invalid method. Choose 'forward', 'backward', or 'central'.")

print("Forward Difference:", num_diff(f, x, h, 'forward'))
print("Backward Difference:",num_diff(f, x, h, 'backward'))
print("Central Difference:", num_diff(f, x, h, 'central'))
```
