```python
# Q 03 ii) Trapezoidal Rule 
```

```python
import numpy as np

def trapz(f, a, b, n):
    if n <= 0:
        print("n must be a positive integer.")

    h = (b - a) / n
    
    x = np.linspace(a, b, n + 1)
    
    y = f(x)
    
    integral = (y[0] + y[-1]) / 2.0
    
    integral += np.sum(y[1:-1])
    
    integral *= h
    
    return integral

def my_f(x_val):
    return 2 * np.pi * (x_val**2.5)

a = 0
b = 2
n = 200

calc = trapz(my_f, a, b, n)

actual = (2 * np.pi * (b**3.5)) / 3.5

print("Calculated integral:",calc)
print("Actual integral: ",actual)

err = calc - actual
print("Integration error:",err)
```

```python
import numpy as np

def f(x):
    return 2 * np.pi * x**2 * np.sqrt(x)

def trap(f, a, b, h):
    x = np.arange(a, b + h, h)
    y = f(x)
    return h * (np.sum(y) - 0.5 * (y[0] + y[-1]))

r = 2
exact = 2 * np.pi * r**3 / 3
hs = [0.5, 0.1, 0.05, 0.01, 0.001]

for h in hs:
    approx = trap(f, 0, r, h)
    err = abs((approx - exact) / exact) * 100
    print(h, approx, exact, err)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def v(t):
    return np.where(t < 2, 2 * t - 1, 3)

def trap(f, a, b, h):
    x = np.arange(a, b + h, h)
    y = f(x)
    return h * (np.sum(y) - 0.5 * (y[0] + y[-1]))

t = np.linspace(0, 4, 400)
plt.plot(t, v(t))
plt.xlabel("t")
plt.ylabel("v(t)")
plt.title("v vs t")
plt.grid(True)
plt.show()

s = 8
hs = [0.5, 0.1, 0.05, 0.01, 0.001]

for h in hs:
    a = trap(v, 0, 4, h)
    e = abs((a - s) / s) * 100
    print(f"h = {h}, s = {a:.6f}, exact = {s:.6f}, error = {e:.6f}")
```
