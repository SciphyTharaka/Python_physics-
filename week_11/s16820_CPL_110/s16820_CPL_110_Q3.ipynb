```python
# Q3 calculate the value of Pi using Monte Carlo method
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of random points (change this value to see accuracy)
N = 10000

# Generate N random points (x, y) in the square [0, 1] x [0, 1]
x = np.random.rand(N)
y = np.random.rand(N)

# Check if each point is inside the quarter circle (x^2 + y^2 <= 1)
inside = x**2 + y**2 <= 1

# Estimate pi using the ratio of points inside the circle
pi_estimate = 4 * np.sum(inside) / N

# Plotting
plt.figure(figsize=(6,6))
plt.scatter(x[inside], y[inside], color='blue', s=1, label='Inside Circle')
plt.scatter(x[~inside], y[~inside], color='red', s=1, label='Outside Circle')
plt.title(f'Estimating Pi using Monte Carlo Method\nEstimated Pi ≈ {pi_estimate:.5f}, N = {N}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axis('square')
plt.show()
```
