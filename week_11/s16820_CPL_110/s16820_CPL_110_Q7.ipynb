```python
import numpy as np
import matplotlib.pyplot as plt

input_voltage = 10
resistor1_nominal = 10000
resistor2_nominal = 2000

tolerance_r1 = 0.05
tolerance_r2 = 0.02

simulations = 100000

r1_min = resistor1_nominal * (1 - tolerance_r1)
r1_max = resistor1_nominal * (1 + tolerance_r1)
r2_min = resistor2_nominal * (1 - tolerance_r2)
r2_max = resistor2_nominal * (1 + tolerance_r2)

r1_samples = np.random.uniform(r1_min, r1_max, simulations)
r2_samples = np.random.uniform(r2_min, r2_max, simulations)

v_out_samples = input_voltage * (r2_samples / (r1_samples + r2_samples))

plt.figure(figsize=(10, 6))
plt.hist(v_out_samples, bins=50, edgecolor='navy', alpha=0.8, color='lightgreen')
plt.title(f'Monte Carlo Voltage Divider Output (Samples = {simulations})', fontsize=14)
plt.xlabel('Output Voltage (V)')
plt.ylabel('Count')
plt.grid(True, linestyle='-.', alpha=0.7)
plt.axvline(np.mean(v_out_samples), color='darkred', linestyle='dotted', linewidth=2,
            label=f'Mean Output Voltage: {np.mean(v_out_samples):.4f} V')
plt.legend()
plt.tight_layout()
plt.show()
```
