```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
nominal_resistance = 470  # ohms
tolerance = 0.05  # ±5%
voltage = 12  # volts
num_resistors = 1000

# 1. Generate resistor values with 5% tolerance (uniform distribution)
min_resistance = nominal_resistance * (1 - tolerance)
max_resistance = nominal_resistance * (1 + tolerance)
resistors = np.random.uniform(min_resistance, max_resistance, num_resistors)

# 2. Calculate current for each resistor using Ohm's law: I = V / R
currents = voltage / resistors

# 3. Plot histogram of resistance values
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(resistors, bins=30, edgecolor='black', color='orange')
plt.title('Resistance Distribution (470Ω ±5%)')
plt.xlabel('Resistance (Ω)')
plt.ylabel('Number of Resistors')
plt.grid(True)

# 4. Plot histogram of current values
plt.subplot(1, 2, 2)
plt.hist(currents, bins=30, edgecolor='black', color='skyblue')
plt.title('Current Through Resistors (12V)')
plt.xlabel('Current (A)')
plt.ylabel('Number of Resistors')
plt.grid(True)

plt.tight_layout()
plt.show()
```
