```python
import numpy as np
import matplotlib.pyplot as plt

# Number of random points
N = 100000

# Generate random (x, y) points in the square [-10, 10] × [-10, 10]
x = np.random.uniform(-10, 10, N)
y = np.random.uniform(-10, 10, N)

# Apply the inequality condition
left_side = (x**2 + y**2 - 2*x)**2
right_side = 4 * (x**2 + 4 * y**2)
inside = left_side <= right_side

# Estimate area
square_area = (20 * 20)  # total area of [-10,10] × [-10,10]
points_inside = np.sum(inside)
area_estimate = square_area * (points_inside / N)

# Plotting
plt.figure(figsize=(7,7))
plt.scatter(x[inside], y[inside], color='green', s=0.5, label='Inside Region')
plt.scatter(x[~inside], y[~inside], color='red', s=0.5, label='Outside Region')
plt.title(f'Estimated Area ≈ {area_estimate:.2f} using Monte Carlo\nPoints: {N}')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('square')
plt.grid(True)
plt.legend()
plt.show()

# Print result
print(f"Estimated area of the region: {area_estimate:.4f}")
```
