```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers uniformly distributed between -1 and 1
x = np.random.uniform(-1, 1, 1000)

# Plot histogram
plt.hist(x, bins=30, edgecolor='black')
plt.title('Histogram of Uniformly Distributed Random Numbers (-1 to 1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()



```
